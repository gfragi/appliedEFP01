ÎœÎ­ÏÎ¿Ï‚ Ï„Î·Ï‚ Î»ÎµÎ¹Ï„Î¿Ï…ÏÎ³Î¯Î±Ï‚ Î­Ï‡ÎµÎ¹ Î®Î´Î· Î¿Î»Î¿ÎºÎ»Î·ÏÏ‰Î¸ÎµÎ¯ Î³Î¹Î± ÎµÏƒÎ¬Ï‚Î§ÏÎµÎ¹Î¬Î¶ÎµÏ„Î±Î¹ Î¼ÏŒÎ½Î¿ Î½Î± Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î®ÏƒÎµÏ„Îµ Ï„Î¿ `list_likes` Î³Î¹Î± Î½Î± Ï…Ï€Î¿Î»Î¿Î³Î¯ÏƒÎµÏ„Îµ Ï„Î¿ `percentage_like`.

```python

def percentage_liked(ratings):
    list_liked = [i>=4 for i in ratings]
    # TODO: Complete the function
    percentage_liked = sum(list_liked)/(len(ratings))
    return percentage_liked

# Do not change: should return 0.5
percentage_liked([1, 2, 3, 4, 5, 4, 5, 1])
```

- ÎµÎ¬Î½ `yrs_ago = 1`, Î· ÏƒÏ…Î½Î¬ÏÏ„Î·ÏƒÎ· Î½Î± ÎµÏ€Î¹ÏƒÏ„ÏÎ­ÏˆÎµÎ¹ Î¼Î¹Î± Ï„Î¹Î¼Î® Ï€ÎµÏÎ¯Ï€Î¿Ï… `0,036`. Î‘Ï…Ï„ÏŒ Î±Î½Ï„Î¹ÏƒÏ„Î¿Î¹Ï‡ÎµÎ¯ ÏƒÎµ Ï€Î¿ÏƒÎ¿ÏƒÏ„ÏŒ Î±ÏÎ¾Î·ÏƒÎ·Ï‚ Ï€ÎµÏÎ¯Ï€Î¿Ï… 3,6%, Ï…Ï€Î¿Î»Î¿Î³Î¹Î¶ÏŒÎ¼ÎµÎ½Î¿ Ï‰Ï‚ (2001078 - 1930992)/1930992.

ÎŸ ÏƒÏ…Î½Î¬Î´ÎµÎ»Ï†ÏŒÏ‚ ÏƒÎ±Ï‚ ÏƒÎ±Ï‚ Î­ÏƒÏ„ÎµÎ¹Î»Îµ Î­Î½Î± Ï€ÏÎ¿ÏƒÏ‡Î­Î´Î¹Î¿ Î¼Î¹Î±Ï‚ ÏƒÏ…Î½Î¬ÏÏ„Î·ÏƒÎ·Ï‚, Î±Î»Î»Î¬ Î´ÎµÎ½ Ï†Î±Î¯Î½ÎµÏ„Î±Î¹ Î½Î± ÎºÎ¬Î½ÎµÎ¹ Ï„Î¿Î½ ÏƒÏ‰ÏƒÏ„ÏŒ Ï…Ï€Î¿Î»Î¿Î³Î¹ÏƒÎ¼ÏŒ.
