This code plots a simple line graph using matplotlib.


```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("My first plot")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.figure(1) # optional
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # red dots

plt.figure(2) # optional
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b*') # blue stars
```

This code creates a figure with two subplots using Matplotlib.

In the first subplot (subplot 121), it plots red dots using the `plot` function.
The x-axis values are [1, 2, 3, 4] and the y-axis values are [1, 4, 9, 16].

In the second subplot (subplot 122), it plots blue stars using the `plot` function.
The x-axis values and y-axis values are the same as in the first subplot.

Note: The `plt.figure` call is optional and can be used to create a new figure if needed.

```python
plt.figure # optional

plt.subplot(121) # 2 rows, 1 column, 1st plot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # red dots

plt.subplot(122) # 2 rows, 1 column, 2nd plot 
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b*') # blue stars
```
